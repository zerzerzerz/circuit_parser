```python
import torch
import dgl
import networkx
from matplotlib import pyplot as plt
```

```python
g = dgl.load_graphs('res/graph.bin')[0][0]
g
```

```python
cell_src, cell_dst = g.edges(etype='cell_out')
net_src, net_dst = g.edges(etype='net_out')
```

```python
edges = []
for src, dst in zip(cell_src, cell_dst):
    edges.append((src.item(), dst.item()))
for src, dst in zip(net_src, net_dst):
    edges.append((src.item(), dst.item()))
```

```python
edges[0]
```

```python
len(edges)
```

```python
g1 = networkx.DiGraph(edges)
g1
```
